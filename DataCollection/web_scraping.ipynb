## Web Scraping Using Python Scripts:

```python
import requests
```

```python
URL = "https://www.scrapethissite.com/pages/simple/"
```

```python
res = requests.get(URL)
```

```python
res ## return status code
```

```python
## print the status code for an http get request
print(res.status_code)
```

```python
## return HTML content of the page
print(res.text)
```

```python
## prints raw bytes of the response
print(res.content)
```

```python
## prints Metadata like content type, server info etc
print(res.headers) 
```
