```python
## import matplotlib library
import matplotlib.pyplot as plt

oscar_winning_movies = ['Titanic', 'The Godfather', 'Forrest Gump', 'Gladiator', 
 'The Lord of the Rings: The Return of the King', 'Parasite', 
 'Slumdog Millionaire', 'Oppenheimer']

oscar_not_winning_movies = ['Inception', 'Interstellar', 'The Dark Knight', 'Avatar', 
 'Dangal', 'Bahubali: The Beginning', '3 Idiots', 'Joker']

year = [
    1972,  
    1994,  # Forrest Gump
    1997,  # Titanic
    2000,  # Gladiator
    2003,  # The Lord of the Rings: The Return of the King
    2008,  # Slumdog Millionaire
    2019,  # Parasite
    2023   # Oppenheimer
]

plt.plot(year, oscar_winning_movies, "go--", label = "Oscar Winning Movies")
plt.plot(year, oscar_not_winning_movies, "*r--",label = "Oscar Not Winning Movies")
plt.title("Oscar Awards")
plt.xlabel("Year")
plt.ylabel("Oscar Awards")
plt.legend(loc = "upper left")
plt.show()
```

```python
## Format String -->> format string is used to style the line on the plot
## there are three parts of the format string  -->> [coloure/marker/line]
## color -> used to given colour to the line, [colour = "red/r"]
## marker -> used to given notation to the coordinates on the line, [marker = ["o"] (circle)
## line -->> used to style the line format

# ex : format string = "r*--" (red color, star format coordinates and dashed lines)

oscar_revenue = [1021,2343,3454,2354,6758]
non_oscar_revenue = [3021,1343,3454, 9354,3758]
year = [2000, 2010, 2020, 2030, 2040]

plt.plot(year, oscar_revenue, "*--r", label = "Oscar Winning Movies")
plt.plot(year, non_oscar_revenue, "s--b", label = "Oscar Not Winning Movies")
plt.title("Oscar Vs Non Oscar Winning Movies")
plt.legend()
plt.show()
```
