```python
## Here are some numpy array inbuilt function -->>
## Aggregate Function:

## np.sum() -->> when we pass the numpy array in the function it will give sum of all elements 
## np.prod() -->> when we pass the numpy array in the function it will give product of all elements
## np.min() -->> when we pass the numpy array in the function it will give minimum element
## np.max() -->> when we pass the numpy array in the function it will give maximum element
## np.mean() -->> when we pass the numpy array in the function it will give mean value
## np.median() -->> when we pass the numpy array in the function it will give median value
## np.std() -->> when we pass the numpy array in the function it will give standard deviation of the array elements
## np.var() -->> when we pass the numpy array in the function it will give variance of the array elements
## np.argmin() -->> when we pass the numpy array in the function it will give index of the minimum element
## np.argmax() -->> when we pass the numpy array in the function it will give index of the maximum element
```

```python
import numpy as np
arr = np.array([1,2,3,4,5])
add =  np.sum(arr) 
product  = np.prod(arr) 
minimum = np.min(arr) 
maximum = np.max(arr) 
mean = np.mean(arr) 
median = np.median(arr)
std_deviation = np.std(arr)
variance = np.var(arr) 
min_index = np.argmin(arr) 
max_index = np.argmax(arr)

print(add)
print(product)
print(minimum)
print(maximum)
print(mean)
print(median)
print(std_deviation)
print(variance)
print(min_index)
print(max_index)
```
