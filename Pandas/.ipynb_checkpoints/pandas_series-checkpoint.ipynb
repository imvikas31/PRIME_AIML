```python
import pandas as pd
```

```python
## creating pandas series -->>
s1 = pd.Series([1,2,3,4,5])
print(s1)
print(type(s1))  ## type of the series
print(s1.dtype)  ## type of data series holds
print(s1.shape)  ## prints number of rows
```

```python
## printing labels of s1 series -->>
print(f"For s2 series, the labels are :\n{s1.index}")
```

```python
## Writing costum labels -->>
s2  =pd.Series([21,23,43,54,65], index = ["Vikas","Yuvraj","Ran","Shyam","Mohan"])
print(s2)

print(f"For s2 series, the labels are:\n{s2.index}")
```

```python
## Properties of Series In Pandas -->>
## Homogenoues -> contains single data type elements, if holds diff data types then type of series is object
## supports vectorized operations -->
s3 = pd.Series([1,2,3,4,5])
s4 = pd.Series([10,20,30,40,50])
print(f"s3 series is :\n{s3}")
print(f"s4 series is :\n{s4}")
print(f"After vectorized addition of s3 and s4 :\n {s3+s4}")

## Sereis are mutable, we can change values of the serise at any label -->>
s3[0] = 100
print(f"After changing the value of s3 seris at label {0} s3 is :\n{s3}")

## we can also drop any value from the series but after droping we have to store the series in new series -->>

changed_s3 = s3.drop(0)
print(f"s3 series after droping the value at label 0 :\n{changed_s3}")
```
