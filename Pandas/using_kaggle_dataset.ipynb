## Handling Missing Values With Pandas

```python
## Handling Missing Values using Pandas -->>
df = pd.read_csv("raw_data.csv")
df
```

```python
## df.isnull() --> prints the data in table form and wherever the data is NaN, NaN appear as True and non null data appears as false
df.isnull() ## df.isna() similar to isnull()
```

```python
## finding number of null/missing values for per/each culumns -->>
df.isnull().sum()
```

```python
## To drop all the null/missing vlues from the dataset -->> df.dropna() -> drops the null values and print data with no null values
## if any row as null values then df.dropna() drops the complete row.
df.dropna()  ## drops rows
df.dropna(axis = 1)  ## drops columns with null/missing values
```

## Handling Duplicate Values with Pandas :

```python
## to find the duplicated values - use .duplicated() method
duplicate_data = df.duplicated()
print(duplicate_data)

## True value in the output shows that the entire row at the corresponfing label is duplicated.

## we can check duplicate for combination of rows or columns 
df[["country","name"]].duplicated()
```

```python
## to delete duplicates we use .drop_duplicates() method
df.drop_duplicates() ## change does not reflect in the original data

## if we wanted to reflect the changes made in the original data then pass an attribute - "inplace =True" in the drop_duplicates() methohd
df.drop_duplicates(inplace=True)
df
```

## Handling Strings In Pandas

```python
## converting lower case to upper
df["gender"].str.lower()
## converting upper case to lower
df["gender"].str.upper()

## capitalize the first letter 
df["gender"].str.capitalize()

## split on basis of space(" ")
df["name"].str.split(" ")

## we can also pass seperator in the split() method -->>
## if we want to split at @ then we can pass @ as str.split("@")

## checking if a column contains any perticular value or not -->>
df["country"].str.contains("US") ## passed parameter is case senstive

## if we want the check irrrespective of case
df["country"].str.contains("india", case=False)
```
