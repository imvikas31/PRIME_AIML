## In this assignment, we have to make the data clean, structured and consistent.
## If you observe the json format store data, you will get to see some inconsistencies like missing age got some user, differed datatype of ratings, duplicate data etc


```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("store_data.json")
print(data)
```

```python
## clean & structure the data - here we will be cleaning the rating part 

def clean_data(data):
    text_to_num = {"one" : 1,"two" : 2,"three" : 3,"four" : 4,"five" : 5}
    cleaned_data = []
    unique_users = set()
    for user in data:
        raw_rating = str(user["rating"]).strip().lower()
        
        
        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]
            
        user["rating"] = raw_rating
        # print(user)

        # Handle missing value --->>
        
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None

        # Handle Duplicate Values --->>
        if(user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
        cleaned_data.append(user)

    return cleaned_data
```

```python
data = clean_data(data)
```

```python
def get_insights(data):
    total_rating = 0
    avg_rating = 0
    poor_rating = 0
    for user in data:
        # find average ratings
        total_rating = total_rating + float(user["rating"])
        avg_rating = total_rating/len(data)

        # find user with ratings > 3.0
        if(float(user["rating"]) < 3.0):
            poor_rating += 1
    print(f"Average rating is : {avg_rating}")
    print(f"Poor rating is : {(poor_rating/len(data))*100} %")
```

```python
get_insights(data)
```

```python
def recommend (data):
    recommendation = []
    for user in data:
        current_recomm = {}
        current_recomm["name"] = user["name"]
        current_recomm["rating"] = user["rating"]
        if (float(user["rating"]) >= 4.0):
            current_recomm["brand"] = "Apple"
        else:
            current_recomm["brand"] = "Samsung"

        recommendation.append(current_recomm)
    return recommendation
            
```

```python
recommend(data)
```
