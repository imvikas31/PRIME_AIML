# Getting Meaningful Insights from the data

```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
```

```python
data = load_data("store_data.json")
```

```python

def get_insights(data):
    total_rating = 0
    avg_rating = 0
    for user in data:
        # find average ratings
        total_rating = total_rating + user["rating"]
        avg_rating = total_rating/len(data)
    print(f"Average rating is : {avg_rating}")
    
```

```python
get_insights(data)
```
